#Get data from all the links

```python
all_data = []
i = 1

for link in all_links:
    response = requests.get(link, impersonate="chrome")
    soup = BeautifulSoup(response.text,'html')

    name_tag = soup.body.find_all('p', attrs={'data-type': 'address'})
    if name_tag:
       name = [title.text.strip() for title in name_tag[0]]
    else:
       name = None

    address_tag = soup.body.find_all('p', attrs={'data-type': 'address'})
    address = [title.text.strip() for title in address_tag[1:]] if address_tag else None

    phone_tag = soup.body.find_all('a', class_ = 'plausible-custom-event-CLICK_TO_CALL')
    phone = [title.text.strip() for title in phone_tag]

    all_data.append({'Company Name' : name, 'Number' : phone, 'Address' : address})
    print(f'Scraped Link {i}')

    i = i + 1

    time.sleep(30)
```
